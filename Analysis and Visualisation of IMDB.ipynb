## Define a column:Rating Category

```python
#define rating function
def define_rating(x):
    if x >= 8.0:
        return "Excellent"
    elif x >= 4.0 and x < 8.0:
        return "Good"
    else:
        return "Poor"
```

```python
#apply function to rating column
movies_df["Rating Category"] = movies_df["Rating"].apply(define_rating)
```

# Top ten(10) Revenue Generated by Each Genre

```python
movies_df.loc[movies_df["Genre"] ==0]
```

```python
Revenue_by_Genre = movies_df.sort_values(by='Revenue', ascending=False).head(10)[['Genre', 'Revenue']]
name= Revenue_by_Genre['Genre']
value= Revenue_by_Genre['Revenue']
plt.figure(figsize=(12,5))
sns.barplot(x= name, y= value)
plt.title('Top 10 Genre by Revenue')
plt.xlabel('Genre')
plt.ylabel('Revenue')
plt.show()
```

# What Movies Generated the Highest Revenue

```python
highest_revenue =movies_df.groupby(["Revenue"], as_index=False)['Movies_Title'].value_counts()[:10].sort_values(by='Movies_Title', ascending=False)
sns.barplot(x= 'Revenue', y= 'Movies_Title', data= highest_revenue);
```

# Top 10 Highest Revenue Generated Movies

```python
Revenue_by_Movies = movies_df.sort_values(by='Revenue', ascending=False).head(15)[['Movies_Tit, 'Revenue']]
name= Revenue_by_Movies['Movies_Title']
value= Revenue_by_Movies['Revenue']
plt.figure(figsize=(12,5))
sns.barplot(x= value, y= name)
plt.title('Top 10 Movies by Revenue')
plt.xlabel('Revenue')
plt.ylabel('Movies_Title')
plt.show()
```

# Top 10 Highest Revenue Generated by Directors

```python
Revenue_by_Director = movies_df.sort_values(by='Director', ascending=False).head(15)[['Director', 'Revenue']]
name= Revenue_by_Director['Director']
value= Revenue_by_Director['Revenue']
plt.figure(figsize=(12,5))
sns.barplot(x= name, y= value)
plt.title('Top 10 Revenue by Director')
plt.xlabel('Director')
plt.ylabel('Revenue')
plt.show()
```

# Average Revenue By Year

```python
highest_revenue =movies_df.groupby(["Year"], as_index=False)['Revenue'].agg('mean').sort_values(by='Year', ascending=False)
plt.figure(figsize=(12,5))
sns.barplot(x= 'Year', y= 'Revenue', data= highest_revenue)
plt.title('Average Revenue By Year')
plt.xlabel('Year')
plt.ylabel('Revenue')
plt.show()
```

# Relationship of Revenue to Rating

```python
x = movies_df.Revenue
y = movies_df.Rating

plt.figure(figsize=(12,5))
plt.scatter(x,y, c="r", marker="o", alpha=0.5)
plt.title("Relationship between Revenue and Rating")
plt.xlabel("Revenue")
plt.ylabel("Rating")

plt.show()
```
